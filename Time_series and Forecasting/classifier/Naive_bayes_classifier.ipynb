This is a code for Naives Bayes Classifier.
The Dataset is included below.

```python
import pandas as pd
```

```python
data = [        
    ["Sunny", "Hot", "High", "Weak", "No"],
    ["Sunny", "Hot", "High", "Strong", "No"],
    ["Overcast", "Hot", "High", "Weak", "Yes"],
    ["Rain", "Mild", "High", "Weak", "Yes"],
    ["Rain", "Cool", "Normal", "Weak", "Yes"],
    ["Rain", "Cool", "Normal", "Strong", "No"],
    ["Overcast", "Cool", "Normal", "Strong", "Yes"],
    ["Sunny", "Mild", "High", "Weak", "No"],
    ["Sunny", "Cool", "Normal", "Weak", "Yes"],
    ["Rain", "Mild", "Normal", "Weak", "Yes"],
    ["Sunny", "Mild", "Normal", "Strong", "Yes"],
    ["Overcast", "Mild", "High", "Strong", "Yes"],
    ["Overcast", "Hot", "Normal", "Weak", "Yes"],
    ["Rain", "Mild", "High", "Strong", "No"]
]

df = pd.DataFrame(data, columns=["Outlook", "Temperature", "Humidity", "Wind", "PlayTennis"])
print(df)
```

```python
X = df[["Outlook", "Temperature", "Humidity", "Wind"]]
y = df["PlayTennis"]

print("Features (X):\n", X)
print("Target (y):\n", y)
```

```python
# prior probabliility
prior_yes = len(df[y == "Yes"]) / len(df)
prior_no = len(df[y == "No"]) / len(df)

print(f"Prior Probability P(Yes): {prior_yes}")
print(f"Prior Probability P(No): {prior_no}")
```

```python
# Conditional Probability P(X|Y)
def conditional_probability(feature, value, target):
    subset = df[y == target]
    count = len(subset[subset[feature] == value])
    return count / len(subset)
```

```python
test_instance = {"Outlook": "Sunny", "Temperature": "Cool", "Humidity": "High", "Wind": "Strong"}

p_yes = prior_yes
p_no = prior_no

for feature, value in test_instance.items():
    prob_yes = conditional_probability(feature, value, "Yes")
    prob_no = conditional_probability(feature, value, "No")
    p_yes *= prob_yes
    p_no *= prob_no
print(f"Posterior Probability P(Yes|X): {p_yes}")
print(f"Posterior Probability P(No|X): {p_no}")
```

```python
# Normalization
total = p_yes + p_no
p_yes_norm = p_yes / total
p_no_norm = p_no / total
print(f"Normalized P(Yes|X): {p_yes_norm}")
print(f"Normalized P(No|X): {p_no_norm}")
```

Based on the output of this dataset, the classification is to not play tennis
